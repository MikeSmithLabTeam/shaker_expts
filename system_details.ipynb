## Creating intruders

These are designed using Openscad. 

`C:\Users\ppzmis\OneDrive - The University of Nottingham\Documents\MikeSmithLabSharedFolder\Fabrication\3DPrinterDesigns\Intruder_openscad`

There is a module called `shaker_objects.scad` which contains various functions to help with the assembly. 

    - L_D sets the lattice dimple spacing to ball size
    - S_D sets the spacing between balls in the crystal along the apothem direction
    - O_D alters the value of L_D by some factor to ensure the balls on the intruder will not mesh with the lattice for orderphobic. For orderphilic this factor should be set to 1.

Particles are red. Dimples are grey circles. Apothem is distance from point of a triangle to midpoint of opposite side.






We 3D print the intruders using PLA. Red for orderphilic and blue for orderphobic. Use the 3D print settings file stored in the intruders openscad folder in Flashprint to get the best results. The black nitrile balls are then glued in with araldite. 

## Perspex hats

The perspex hats are created in openscad and exported as a dxf file before being transferred to laser cutter software. Both pieces are cut from 3mm thick perspex and then glued together with superglue. Use `Lasersoft` to import dxf and set up the cut. 
2mm/s 25%power and 20%corner power works fine. If not perhaps mirrors are getting dirty.
